```python
def scheduler(df_in, sch,except_sch):
    """Adds a new column to the input dataframe. Takes values from the sch dictionary and assigns 
       them according to the scheduled time (weekly or daily)
    Args:
        dataframe object -> index must be a timestamp with the name 'time'.
        sch -> it can be daily (with one schedule for each day) or weekly (separate schedule for each day)
               the sch dictionary (in case of weekly) must have at lease one schedule for each day.
    """
    # copy imported dataframe to a new variable
    df = df_in.copy()
    # extract column names to be used again later
    column_names = df.columns.values.tolist()
    
    # pop the index ('time' column) out and make it to a column
    df.reset_index(inplace=True)
    # Convert 'time' column to datetime (if not already)
    df['time'] = pd.to_datetime(df['time'])
    # Create a new column 'weekday' with respective weekday numbers
    df['weekday'] = df['time'].dt.dayofweek
    # Extract time from the 'time' column and put into new column
    df['time_only'] = pd.to_datetime(df['time'],format= '%H:%M:%S' ).dt.time
    df['date_only'] = pd.to_datetime(df['time'],format= '%H:%M:%S' ).dt.date
    # Extract keys length (would be 7 for weekly and 1 for daily)
    for keys in sch.keys():
        length = len(sch[keys])
    # IN CASE OF WEEKLY SCHEDULE DICTIONARY   
    if length == 7:
        for i in sch.keys():
            key = i
            for j in range(len(sch[i])):
                for k in range(len(sch[i][j])):
                
                # Extract information for respective day and save into variables
                    hour = sch[i][j][k]['hour']
                    minute = sch[i][j][k]['minute']
                    second = sch[i][j][k]['second']
                    value = sch[i][j][k]['value']
                    
                    # first schedule for next day - if day is sunday
                    if j == 6 :
                        hour_n = sch[i][0][0]['hour']
                        minute_n =sch[i][0][0]['minute']
                        second_n =sch[i][0][0]['second']
                    # first schedule for next day - if day is not sunday
                    else:
                        hour_n = sch[i][j+1][0]['hour']
                        minute_n =sch[i][j+1][0]['minute']
                        second_n =sch[i][j+1][0]['second']
                    
                    # create timestamp using info from dictionary
                    time = str(hour)+':'+str(minute)+':'+str(second)
                    # create timestamp for next day from dictionary
                    time_next_day = str(hour_n)+':'+str(minute_n)+':'+str(second_n)
                    # Convert both new 'time' columns to timestamp
                    df['time_new'] = pd.to_datetime(time,format= '%H:%M:%S')
                    df['time_new'] = pd.to_datetime(df['time_new'],format= '%H:%M:%S' ).dt.time
                    df['time_new_next_day'] = pd.to_datetime(time_next_day,format= '%H:%M:%S')
                    df['time_new_next_day'] = pd.to_datetime(df['time_new_next_day'],format= '%H:%M:%S').dt.time
                    # For Sunday
                    if j == 6:
                        # Set value for current day
                        df.loc[((df['time_only'] >= df['time_new']) & (df['weekday'] == j)) , i] = value
                        # overflow the same value from current day to next day until first schedul appears
                        df.loc[((df['time_only'] < df['time_new_next_day']) & (df['weekday'] == 0)) , i] = value
                    # For all other week days
                    else :
                        # Set value for current day
                        df.loc[((df['time_only'] >= df['time_new']) & (df['weekday'] == j)) , i] = value                          
                        # overflow the same value from current day to next day until first schedule appears
                        df.loc[((df['time_only'] < df['time_new_next_day']) & (df['weekday'] == j+1))  , i] = value
    # IN CASE OF DAILY SCHEDULE DICTIONARY
    else:
        for i in sch.keys():
            key=i
            for j in range(len(sch[i])):
                for k in range(len(sch[i][j])):
                    for day in range(0,7):
                        # Extract information for respective day and save into variables
                        hour = sch[i][j][k]['hour']
                        minute = sch[i][j][k]['minute']
                        second = sch[i][j][k]['second']
                        value = sch[i][j][k]['value']
                        # first schedule for next day
                        hour_n = sch[i][0][0]['hour']
                        minute_n = sch[i][0][0]['minute']
                        second_n = sch[i][0][0]['second']
                        # create timestamp using info from dictionary
                        time = str(hour)+':'+str(minute)+':'+str(second)
                        # create timestamp for nex day using info from dictionary
                        time_next_day = str(hour_n)+':'+str(minute_n)+':'+str(second_n)
                        # Convert both new 'time' columns to timestamp
                        df['time_new'] = pd.to_datetime(time,format= '%H:%M:%S')
                        df['time_new'] = pd.to_datetime(df['time_new'],format= '%H:%M:%S' ).dt.time
                        df['time_new_next_day'] = pd.to_datetime(time_next_day,format= '%H:%M:%S')
                        df['time_new_next_day'] = pd.to_datetime(df['time_new_next_day'],format= '%H:%M:%S').dt.time
                        # For Sunday
                        if day == 6 :
                            # Set value for current day
                            df.loc[((df['time_only'] >= df['time_new']) & (df['weekday'] == day)) , i] = value
                            # overflow the same value from current day to next day until first schedul appears
                            df.loc[((df['time_only'] < df['time_new_next_day']) & (df['weekday'] == 0)) , i] = value
                        # For all other days
                        else :
                            # Set value for current day
                            df.loc[((df['time_only'] >= df['time_new']) & (df['weekday'] == day)) , i] = value
                            # overflow the same value from current day to next day until first schedul appears
                            df.loc[((df['time_only'] < df['time_new_next_day']) & (df['weekday'] == day+1)) , i] = value
            
            
    if except_sch is None:
        return df
    else:
        for i in except_sch.keys():
            date = i
            print(i)
            for j in range(len(except_sch[i])):
                for k in range(len(except_sch[i][j])):
                    
                    hour_ex = except_sch[i][j][k]['hour']
                    minute_ex= except_sch[i][j][k]['minute']
                    second_ex = except_sch[i][j][k]['second']
                    value_ex = except_sch[i][j][k]['value']
                    print(value_ex)
                    
                    time = str(hour_ex)+':'+str(minute_ex)+':'+str(second_ex)
                    df['time_new'] = pd.to_datetime(time,format= '%H:%M:%S')
                    df['time_new'] = pd.to_datetime(df['time_new'],format= '%H:%M:%S' ).dt.time
                    df['date_ex'] = pd.to_datetime(date,format= '%Y-%m-%d')
                    df['date_ex'] = pd.to_datetime(df['date_ex'],format='%Y-%m-%d').dt.date
                    df['next_date'] = df['date_only'] + datetime.timedelta(days=1) 
#                     print(df['date_only'] - (df['date_only']+datetime.timedelta(days=1))==1)
#                     print(df['date_only']==date)
#                     print(date)
                    df.loc[((df['time_only'] >= df['time_new']) & (df['date_only']== df['date_ex'])), df.columns[-3]] = value_ex
#                     df.loc[((df['time_only'] < df['time_new_next_day']) & (df['date_only'] - (df['date_only']+datetime.timedelta(days=1))==1),df.columns[-3])] = value_ex
    # get all the column names and add the new column name that has values
    # new column name is the key of the outer dictionary
    column_names.append(key)
    # Set 'time' column as index again
    df.set_index('time',inplace=True)
    # Return df
    return df
                
            
```
